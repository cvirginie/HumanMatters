```python
import nltk
import pickle
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import re
from nltk.corpus import wordnet as wn
from nltk import word_tokenize, pos_tag
from collections import defaultdict

nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
import os 
import pandas as pd
import re
from datetime import datetime
import numpy as np

# variable definition
path_bulletins = r".\data-science-for-good-city-of-los-angeles\CityofLA\Job Bulletins"
bulletins=os.listdir(path_bulletins)
num = len(bulletins)
text_postings = ""

for no in range(0,num):
    with open(path_bulletins+"/" +bulletins[no],encoding="ISO-8859-1") as f:         #reading files 
        try:
            file=f.read().replace('\t','')
            nblines = file.count('\n')
            raw_text=file.replace('\n',' ')
            text_postings = text_postings + file
        except Exception as e:
            print(e)
```

```python
nltk.download('punkt')

lyrics=text_postings

def stopwords(text):
    return ' '.join([w for w in word_tokenize(text.lower()) if not w in stop_words])

lyrics=stopwords(lyrics)

stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 20,max_words=100).generate(lyrics)
# plot the WordCloud image                        
plt.figure(figsize = (6, 6), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
